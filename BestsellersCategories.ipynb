```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

### Amazon Top 50 Bestselling Books 2009 - 2019
#### About Dataset
Dataset on Amazon's Top 50 bestselling books from 2009 to 2019. Contains 550 books, data has been categorized into fiction and non-fiction using Goodreads

### Campos del dataset
Origen de datos: https://www.kaggle.com/datasets/sootersaalu/amazon-top-50-bestselling-books-2009-2019

|Title | Description | Type |
|------|-------------|------|
|Name | Name of the Book | String
| Author | The author of the Book | String
| User Rating | Amazon User Rating |Number
| Reviews | Number of written reviews on amazon | Number
| Price | The price of the book (As at 13/10/2020) | Number
| Year | The Year(s) it ranked on the bestseller | Number
| Genre | Whether fiction or non-fiction | String

```python
dataBestsellers=pd.read_csv("bestsellers with categories.csv")
dataBestsellers
```

```python
items=["Reviews"]

for f in items:
    fig=plt.figure(figsize=(12,4))
    sb.barplot(x="Year", y=f, data=dataBestsellers)
    plt.xticks(rotation=90)
    plt.grid()
    plt.title(f)
    plt.show()
```

```python
indexYear=[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]

filterNonFiction=dataBestsellers["Genre"]=="Non Fiction"
auxiliaryNonFiction=dataBestsellers[filterNonFiction]
genreNonFiction=auxiliaryNonFiction.groupby("Year")["User Rating"].count()

filterFiction=dataBestsellers["Genre"]=="Fiction"
auxiliaryFiction=dataBestsellers[filterFiction]
genreFiction=auxiliaryFiction.groupby("Year")["User Rating"].count()

generalTable=pd.DataFrame({"Non Fiction":genreNonFiction,"Fiction":genreFiction},index=indexYear)
generalTable.plot(kind="line")##,stacked=True)
plt.show()
```
